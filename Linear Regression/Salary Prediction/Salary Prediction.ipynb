```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
x = np.array(df.YearsExperience).reshape(-1, 1)
y = np.array(df.Salary)
```

```python
model = linear_model.LinearRegression()
model.fit(x, y)
```

```python
plt.scatter(x, y, color="red")
plt.plot(x, model.predict(x), color="blue")
```

```python
years = float(input("Enter years of experience to predict salary: "))
predicted_salary = model.predict([[years]])
f"Predicted salary for {years} years of experience: {predicted_salary[0]:.0f}"  
```

```python
plt.scatter(x, y, color="red")

plt.plot(x, model.predict(x), color="blue")

plt.scatter(years, predicted_salary, color="green", marker="x", s=100, label="Prediction")

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Prediction")
plt.legend()
```
